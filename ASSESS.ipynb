Ques - 1 - : What is Simple Linear Regression (SLR)? Explain its purpose?

Answer - Simple Linear Regression (SLR) is a statistical technique used to study the relationship between two variables, namely one independent variable and one dependent variable. It assumes that the relationship between the variables can be represented by a straight line. The mathematical model of simple linear regression is given by the equation


Y=a+bX, where
𝑌
Y is the dependent variable,
𝑋
X is the independent variable,
𝑎
a is the intercept, and
𝑏
b is the slope of the line. The slope indicates the rate of change of the dependent variable with respect to the independent variable, while the intercept represents the value of the dependent variable when the independent variable is zero. The main purpose of simple linear regression is to establish and quantify the relationship between variables, predict the value of one variable based on the other, and analyze trends in data. It is widely used in economics, engineering, and science for forecasting and decision-making. For example, simple linear regression can be used to predict a student’s exam score based on the number of hours studied.

Answer - The key assumptions of Simple Linear Regression (SLR) are the conditions that must be satisfied for the regression results to be valid and reliable. First, there must be a linear relationship between the independent variable
𝑋
X and the dependent variable
𝑌
Y, meaning that the change in
𝑌
Y is proportional to the change in
𝑋
X. Second, the errors (residuals) are independent, i.e., the error associated with one observation does not influence another. Third, the model assumes homoscedasticity, which means that the variance of the errors remains constant for all values of the independent variable. Fourth, the errors are assumed to be normally distributed with a mean of zero, which is important for hypothesis testing and confidence intervals. Fifth, the independent variable
𝑋
X is assumed to be measured without error or with negligible error. These assumptions ensure accurate estimation, prediction, and inference in simple linear regression. For example, when predicting exam marks based on hours studied, SLR assumes that marks increase linearly with study hours, the variation in marks is consistent across all study times, and other random factors affecting marks are normally distributed.

Ques - 5-  What is the method of least squares in linear regression?


Answer - The method of least squares is a mathematical technique used in linear regression to estimate the parameters of the regression line such that it best fits the given data. In this method, the “best fit” line is obtained by minimizing the sum of the squares of the residuals, where a residual is the vertical difference between an observed value of the dependent variable and the value predicted by the regression line. Squaring the residuals ensures that positive and negative errors do not cancel each other and gives greater weight to larger errors. For a simple linear regression model -

Y=a+bX, the least squares method determines the values of the intercept
𝑎
a and slope
𝑏
b that minimize the total squared error between observed and predicted values. The primary purpose of this method is to obtain the most accurate and reliable estimates of the regression coefficients based on available data. For example, when predicting exam marks from hours studied, the least squares method finds the line that minimizes the overall deviation between actual marks and predicted marks, thereby providing the most suitable linear model for prediction and analysis.t

Ques- 6  What is Logistic Regression? How does it differ from Linear Regression?


Answer - Linear regression is used when the dependent variable is continuous, such as height, weight, or marks, and it assumes a linear relationship between variables, producing outputs that can take any real value. Logistic regression, on the other hand, is used when the dependent variable is categorical (binary) and produces outputs limited between 0 and 1, making it suitable for classification problems. Linear regression uses the least squares method for estimation, while logistic regression uses maximum likelihood estimation. For example, linear regression can be used to predict a student’s exam score based on hours studied, whereas logistic regression can be used to predict whether a student will pass or fail based on the same input. Thus, logistic regression is mainly applied in classification, while linear regression is applied in prediction of continuous values.

Ques- 7 - Name and briefly describe three common evaluation metrics for regression
models

Answer - Mean Absolute Error (MAE):
MAE measures the average of the absolute differences between the actual values and the predicted values. It gives a clear idea of the average prediction error in the same units as the dependent variable and is less sensitive to outliers.

Mean Squared Error (MSE):
MSE is the average of the squared differences between actual and predicted values. By squaring the errors, it penalizes larger errors more heavily, making it useful when large deviations are particularly undesirable.

Coefficient of Determination (R²):
R² indicates the proportion of variance in the dependent variable that is explained by the regression model. Its value ranges from 0 to 1, where a higher value signifies a better fit of the model to the data.

Ques - 8 - What is the purpose of the R-squared metric in regression analysis?

Answer - The purpose of the R-squared (R²) metric in regression analysis is to measure how well a regression model explains the variability of the dependent variable using the independent variable(s). R² represents the proportion of total variation in the dependent variable that is explained by the regression model, and its value ranges from 0 to 1. An R² value close to 1 indicates that the model explains most of the variability in the data, while a value close to 0 indicates that the model explains very little. Thus, R² helps assess the goodness of fit of a regression model and compare the explanatory power of different models on the same dataset. However, R² does not indicate whether the model is appropriate or whether predictions are unbiased; it only reflects how well the model fits the observed data. For example, if a regression model predicting exam marks from hours studied has an R² of 0.85, it means that 85% of the variation in exam marks is explained by study hours, while the remaining 15% is due to other factors or random variation.

Ques - 9 - Write Python code to fit a simple linear regression model using scikit-learn
and print the slope and intercept.

```python
import numpy as np
from sklearn.linear_model import LinearRegression

# Sample data (independent variable X and dependent variable y)
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)  # must be 2D
y = np.array([2, 4, 5, 4, 5])

# Create and fit the model
model = LinearRegression()
model.fit(X, y)

# Print slope and intercept
print("Slope (b):", model.coef_[0])
print("Intercept (a):", model.intercept_)
```
