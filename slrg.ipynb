1. What is Simple Linear Regression?

Ans - Simple Linear Regression is a statistical technique used to study the relationship between one independent variable and one dependent variable. It assumes that the dependent variable Y changes linearly with the independent variable X. The relationship is represented by the equation Y = mX + c, where m is the slope and c is the intercept. The main objective of simple linear regression is to predict the value of Y for a given value of X and to understand how strongly X influences Y. It is widely used in forecasting, business analytics, and scientific research. For example, a teacher may use simple linear regression to predict a student’s marks based on the number of hours studied. If more hours lead to higher marks in a roughly straight-line pattern, SLR becomes an appropriate model.

2. Key Assumptions of Simple Linear Regression?

Ans - Simple Linear Regression is based on several important assumptions. First, there must be a linear relationship between X and Y. Second, the observations should be independent of each other. Third, the variance of errors must be constant, which is called homoscedasticity. Fourth, the residuals should follow a normal distribution. Fifth, the independent variable should be measured without error. If these assumptions are violated, the regression results may be misleading. For example, if residuals show a curved pattern instead of random scatter, the linearity assumption is violated and predictions will not be reliable.

3. Meaning of Coefficient m in Y = mX + c ?

Ans - In the regression equation Y = mX + c, the coefficient m represents the slope of the regression line. It measures how much the dependent variable changes for one-unit change in the independent variable. The sign of m shows the direction of the relationship: positive m indicates direct relation, while negative m shows inverse relation. The magnitude of m tells the strength of impact. For example, if m = 5 in a model predicting salary from years of experience, it means salary increases by 5 units for every additional year of experience.

4. Meaning of Intercept c ?

Ans - The intercept c represents the predicted value of Y when X is zero. It provides the starting point of the regression line on the Y-axis. The intercept may or may not have practical meaning depending on the context. For example, in a model predicting electricity bill from units consumed, c represents the fixed charge when consumption is zero. However, in some cases like predicting marks from study hours, X=0 may not be realistic, so the intercept only serves mathematical purpose.

6. Purpose of Least Squares Method ?

Ans - The least squares method is used to obtain the best regression line by minimizing the sum of squared differences between actual and predicted values. Squaring removes negative signs and gives more weight to large errors. This method produces unbiased and efficient estimates of m and c. For example, in predicting house prices, many lines can be drawn, but least squares selects the line with minimum total prediction error.

7. Interpretation of R² ?

Ans -The coefficient of determination (R²) measures how well the regression model explains the variability of the dependent variable. It ranges from 0 to 1. A higher R² indicates better model fit. R² = 0.75 means 75% of variation in Y is explained by X, while 25% is due to other factors. However, high R² does not prove causation. For example, if R² between rainfall and crop yield is 0.80, rainfall strongly explains yield but other factors like soil also matter.

8. What is Multiple Linear Regression?

Ans - Multiple Linear Regression is an extension of simple regression where more than one independent variable is used to predict a dependent variable. The model is
Y = b0 + b1X1 + b2X2 + … + bnXn.
It helps analyze combined influence of several factors. It is useful when real-life outcomes depend on many variables. For example, a company may predict sales based on price, advertising, and income level simultaneously.

9. Difference between Simple and Multiple Regression ?

Ans - The main difference is the number of predictors. Simple regression uses one independent variable, while multiple regression uses two or more. Interpretation in SLR is easier, but MLR gives more realistic models. MLR can handle complex relationships but requires checking multicollinearity. For example, predicting weight from height is SLR, whereas predicting weight from height, age, and diet is MLR.

10. Assumptions of Multiple Regression ?

Ans - MLR assumes linearity, independence of errors, homoscedasticity, normality of residuals, and absence of multicollinearity among predictors. Violation leads to biased coefficients. For example, if two variables like height and arm length are highly correlated, individual effects cannot be separated.

11. What is Heteroscedasticity?

Ans - Heteroscedasticity occurs when error variance is not constant across observations. It causes unreliable standard errors and hypothesis tests. It can be detected using residual plots or Breusch-Pagan test. For example, in income prediction, variance of errors may increase for higher incomes forming a funnel shape.

12. Improving Model with Multicollinearity?

Ans -Multicollinearity can be reduced by removing correlated variables, combining them, using PCA, or applying Ridge/Lasso regression. These methods stabilize coefficients. For example, instead of using height in cm and inches together, keep only one.

13. Transforming Categorical Variables ?

Ans- Categorical variables must be converted to numerical form using dummy encoding, one-hot encoding, or label encoding. This allows inclusion in regression. For example, gender can be coded as Male=0, Female=1.

14. Role of Interaction Terms ?

Ans - Interaction terms show that the effect of one predictor depends on another. They improve model realism. For example, impact of study hours on marks may depend on teaching method, modeled as Hours*Method.

15. Intercept in SLR vs MLR?

Ans - In SLR, intercept is value of Y when single X=0. In MLR, it is value when all predictors are zero, which may be unrealistic. Hence interpretation in MLR is more cautious.

16. Significance of Slope ?

Ans - Slope indicates direction, strength, and contribution of predictor to outcome. It directly affects predictions. For example, negative slope between price and demand shows that higher price reduces demand.

17. Context from Intercept ?

Ans -Intercept provides baseline level of dependent variable and helps compare groups. In cost models it represents fixed cost, giving managerial insight.

18. Limitations of R² ?

Ans - R² increases with more variables, ignores overfitting, and does not show causal relation or prediction error. Adjusted R² and RMSE should also be checked.

19. Large Standard Error?

Ans -Large standard error means coefficient is unstable and may be insignificant due to small sample or multicollinearity. Predictions become unreliable.

20. Identifying Heteroscedasticity ?

Ans - In residual vs fitted plot, non-constant spread or funnel shape indicates heteroscedasticity. It must be corrected using transformation or robust errors because it affects inference.

21. High R² but Low Adjusted R² ?

Ans - This means many unnecessary predictors are included. Adjusted R² penalizes extra variables, indicating overfitting.

22. Importance of Scaling ?

Ans - Scaling puts variables on same range, improves gradient methods, and is essential for regularized regression. Example: age (0-80) and income (0-100000) must be standardized.

23. What is Polynomial Regression?

Ans - Polynomial regression models nonlinear relationship by including powers of X such as X², X³ while remaining linear in parameters. Example: growth of bacteria over time shows curved pattern.

24. Difference between Polynomial and Linear ?

Ans - Linear regression fits straight line; polynomial fits curves. Polynomial captures complex patterns but risks overfitting. Example: temperature vs electricity load often requires quadratic model.

25. When is Polynomial Regression Used?

Ans - Polynomial regression is used when the relationship between the independent and dependent variables is non-linear and cannot be properly modeled using a straight line. It is suitable when data shows curves, peaks, or diminishing returns. Instead of fitting a simple line, polynomial regression fits a curved function by adding higher-order powers of X. This method is widely applied in engineering, economics, and natural sciences where growth or decay follows nonlinear patterns. For example, the relationship between temperature and electricity demand is often U-shaped: demand increases at very low and very high temperatures. In such cases, a quadratic or cubic polynomial model gives better predictions than simple linear regression.





27. Can Polynomial Regression be Applied to Multiple Variables?

Ans -Yes, polynomial regression can be extended to multiple independent variables, known as multivariate polynomial regression. In this case, polynomial terms are created for each variable as well as interaction terms between variables. The model may include terms like
. This allows the model to capture complex surfaces instead of simple curves. However, the number of terms grows rapidly with degree and variables, increasing computational cost and risk of overfitting. For example, house price can be modeled using size and age with terms such as Size² and Size×Age to capture nonlinear effects.

28. Limitations of Polynomial Regression ?

Ans -Polynomial regression has several limitations. High-degree polynomials may overfit the training data and perform poorly on new data. The model can behave unpredictably outside the observed range (extrapolation problem). Multicollinearity between polynomial terms makes coefficient interpretation difficult. It is also sensitive to outliers and requires careful scaling of variables. Choosing the correct degree is challenging; too low underfits, too high overfits. For example, a 10th-degree polynomial may perfectly pass through all data points but give unrealistic oscillations between them.

29. Methods to Evaluate Model Fit and Select Degree ?

Ans - Several methods are used to choose the appropriate polynomial degree. Cross-validation is the most reliable technique to compare prediction error on unseen data. Adjusted R², AIC, BIC, and RMSE are also used as evaluation metrics. Residual analysis helps check whether systematic patterns remain. The bias-variance trade-off must be considered: increasing degree reduces bias but increases variance. For example, a data scientist may train models with degrees 1 to 5 and select the one with minimum cross-validated RMSE.

30. Importance of Visualization in Polynomial Regression ?

Ans - Visualization is very important in polynomial regression because it helps understand the shape of the relationship and detect overfitting. Scatter plots with fitted curves show whether the chosen degree follows the data realistically. Residual plots reveal heteroscedasticity and outliers. Visualization also helps communicate results to non-technical users. For example, plotting sales against advertising with a quadratic curve may clearly show saturation effect that cannot be seen from coefficients alone.

31 -  How is polynomial regression implemented in Python?

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
import numpy as np

X = np.array([1,2,3,4,5]).reshape(-1,1)
y = np.array([2,6,7,8,10])

poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

model = LinearRegression()
model.fit(X_poly, y)
pred = model.predict(X_poly)
```
